```python
# prompt: save all file path in train directory in train.txt file in train directory, create the file if not exist

import os

target = 'valid'
train_dir = f'/content/Yolo-FastestV2/dataset/{target}'
train_txt_path = os.path.join(train_dir, f'{target}.txt')

with open(train_txt_path, 'w') as f:
  for root, _, files in os.walk(train_dir):
    for file in files:
      if file.endswith('.jpg') or file.endswith('.png'):
        file_path = os.path.join(root, file)
        f.write(file_path + '\n')
```
